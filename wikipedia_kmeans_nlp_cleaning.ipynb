```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# Load or use your dataframe
# df = pd.read_csv("wikipedia_visits.cs")

# Select numeric columns for clustering
X = df.select_dtypes(include=['int64','float64'])

# Create KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
kmeans.fit(X)

# Find closest cluster center and distance for each point
closest, distances = pairwise_distances_argmin_min(X, kmeans.cluster_centers_)

# Store cluster labels
df["cluster"] = closest

# Store distance to cluster center
df["distance"] = distances

# Define outlier threshold (95th percentile)
threshold = df["distance"].quantile(0.95)

# Remove outliers
clean_df = df[df["distance"] < threshold]

# Reset index
clean_df = clean_df.reset_index(drop=True)

print(clean_df.head())
```
